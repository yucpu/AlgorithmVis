```python
# Add any additional libraries or submodules below
# Display plots inline
%matplotlib inline
# Data libraries
# Plotting libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


plt.rcParams['figure.figsize'] = (7,7)
plt.rcParams['figure.dpi'] = 80
plt.rcParams.update({'font.size': 8})
```

```python
qa = pd.read_csv("data.csv");
qa.loc[qa["Major"]=="Yes", "Major"] = "CS_Related"
qa.loc[qa["Major"]=="No", "Major"] = "Not_Related"
qa.loc[qa["Experience"]=="Yes", "Experience"] = "Have"
qa.loc[qa["Experience"]=="No", "Experience"] = "Not Have"
qa.loc[qa["Knowledge"]=="Yes", "Knowledge"] = "Have"
qa.loc[qa["Knowledge"]=="No", "Knowledge"] = "Not Have"
```

```python
qa.loc[qa["MSCheck"]=="Yes", "MSCheck"] = "correct"
qa.loc[qa["MSCheck"]=="No", "MSCheck"] = "wrong"
```

```python
qa.loc[qa["CPPCheck"]=="Yes", "CPPCheck"] = "correct"
qa.loc[qa["CPPCheck"]=="No", "CPPCheck"] = "wrong"
```

```python
qa.loc[qa["IsUseful_MS"]=="Strongly agree", "IsUseful_MS"] = "agree"
qa.loc[qa["IsUseful_MS"]=="Neither agree nor disagree", "IsUseful_MS"] = "neutral"
qa.loc[qa["IsUseful_MS"]=="Strongly disagree", "IsUseful_MS"] = "disagree"
```

```python
qa.loc[qa["IsUseful_CPP"]=="Strongly agree", "IsUseful_CPP"] = "agree"
qa.loc[qa["IsUseful_CPP"]=="Neither agree nor disagree", "IsUseful_CPP"] = "neutral"
qa.loc[qa["IsUseful_CPP"]=="Strongly disagree", "IsUseful_CPP"] = "disagree"
```

```python
qa.loc[qa["GoodUI"]=="Yes", "GoodUI"] = "good UI"
qa.loc[qa["GoodUI"]=="No", "GoodUI"] = "bad UI"
```

```python
qa.loc[qa["isTrouble"]=="Yes", "isTrouble"] = "Have Trouble"
qa.loc[qa["isTrouble"]=="No", "isTrouble"] = "No Trouble"
```

```python
qa.to_csv('newData.csv',index=False)
```

```python
qa.info()
```

```python
fig, axs = plt.subplots(nrows=3,ncols=2)

cols = ["CS_Related", "Not_Related"];
axs[0,0].set_title("CS_Related");
axs[0,1].set_title("Not_Related");

sectionOne = qa[qa['Major']=="CS_Related"]
sectionTwo = qa[qa['Major']=="Not_Related"]


CheckCS = sectionOne[['MSCheck',"CPPCheck"]];
sns.countplot(x="variable", hue="value", hue_order=['correct', 'wrong'] ,data= CheckCS.melt(), ax=axs[0,0])

CheckNCS = sectionTwo[['MSCheck',"CPPCheck"]];
sns.countplot(x="variable", hue="value", hue_order=['correct', 'wrong'] , data= CheckNCS.melt(), ax=axs[0,1])

UsefulCS = sectionOne[["IsUseful_MS","IsUseful_CPP"]]
sns.countplot(x="variable", hue="value", hue_order=['agree', 'neutral', 'disagree'] , data= UsefulCS.melt(), ax=axs[1,0])

UsefulNCS = sectionTwo[["IsUseful_MS","IsUseful_CPP"]]
sns.countplot(x="variable", hue="value", hue_order=['agree', 'neutral', 'disagree'] ,data= UsefulNCS.melt(),ax=axs[1,1])

HLCS = sectionOne[['MSHLControl',"CPPHLControl"]]
HLNCS = sectionTwo[['MSHLControl',"CPPHLControl"]]
sns.countplot(x="variable", hue="value",hue_order=['Agree', "Disagree"] , data= HLCS.melt(), ax=axs[2,0])

sns.countplot(x="variable", hue="value",hue_order=['Agree', "Disagree"] , data= HLNCS.melt(), ax=axs[2,1])
```
